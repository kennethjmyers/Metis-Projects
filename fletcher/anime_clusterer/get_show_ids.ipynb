```python
import pandas as pd
```

```python
df = pd.read_csv('prescraped_list.csv', names=['url', 'airstatus', 'rating', 'rank', 'popularity', 'members', 'favorites'])
```

```python
split = list(df['url'].str.split('/', 3))
```

```python
ids = [x[2] for x in split]
names = [x[3].replace('_', ' ') for x in split]
df['id'] = pd.Series(ids)
df['name'] = pd.Series(names)
df = df[['id', 'name', 'rating', 'rank', 'popularity', 'members', 'favorites']]
```

```python
df['rank'] = df['rank'].str.replace('#', '')
df['popularity'] = df['popularity'].str.replace('#', '')
```
